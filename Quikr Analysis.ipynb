## Problems found in data

- names are pretty inconsistent
- names have company names attached to it
- some names are spam like 'Maruti Ertiga showroom condition with' and 'Well mentained Tata Sumo'
- company: many of the names are not of any company like 'Used', 'URJENT', and so on.
- year has many non-year values
- year is in object. Change to integer
- Price has Ask for Price
- Price has commas in its prices and is in object
- kms_driven has object values with kms at last.
- It has nan values and two rows have 'Petrol' in them
- fuel_type has nan values

## Cleaning Data 

#### 1) year has many non-year values

```python
car=car[car['year'].str.isnumeric()]
```

#### 2) year is in object. Change to integer

```python
car['year']=car['year'].astype(int)
```

#### 3) Price has Ask for Price

```python
car=car[car['Price']!='Ask For Price']
```

#### 4) Price has commas in its prices and is in object

```python
car['Price']=car['Price'].str.replace(',','').astype(int)
```

#### 5) kms_driven has object values with kms at last.

```python
car['kms_driven']=car['kms_driven'].str.split().str.get(0).str.replace(',','')
```

#### 6) It has nan values and two rows have 'Petrol' in them

```python
car=car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
```

#### 7) fuel_type has nan values

```python
car=car[~car['fuel_type'].isna()]
```

```python
car.shape
```

#### 8) Company does not need any cleaning now. Changing car names. Keeping only the first three words

```python
car['name']=car['name'].str.split().str.slice(start=0,stop=3).str.join(' ')
```

#### 9) Resetting the index of the final cleaned data

```python
car=car.reset_index(drop=True)
```

### Extracting Training Data

We want price of the car. And here Price is depended on 'name','company', 'year when car bought', 'kms driven' and what is the fuel type

```python
X=car[['name','company','year','kms_driven','fuel_type']]
y=car['Price']
```

```python
X
```

```python
y.shape
```

### Applying Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

#### Creating an OneHotEncoder object to contain all the possible categories

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

#### Creating a column transformer to transform categorical columns

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

#### Linear Regression Model

```python
lr=LinearRegression()
```

#### Making a pipeline

```python
pipe=make_pipeline(column_trans,lr)
```

#### Fitting the  model

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

#### Checking R2 Score

```python
r2_score(y_test,y_pred)
```

#### Finding the model with a random state of TrainTestSplit where the model was found to give almost 0.89 as r2_score

Our data is very very small sized. There are only around 800 rows. so there are major difference in r2 score. So we need to get maximum r2 score value.

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
pipe.predict(pd.DataFrame(columns=X_test.columns,data=np.array(['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']).reshape(1,5)))
```

```python
pipe.predict(pd.DataFrame(columns=X_test.columns,data=np.array(['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']).reshape(1,5)))
```

#### The best model is found at a certain random state 

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
